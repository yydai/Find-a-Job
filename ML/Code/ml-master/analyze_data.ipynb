### 6.属性间相关性

```python
import pandas as pd

pd.set_option('display.width',100)  ##default 80
pd.set_option('display.precision', 2) ##default 3

filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pd.read_csv(filename, delimiter=',', header=None, names=names)

cor = dataframe.corr(method='pearson')  #correlation 相关性
print (cor)

cov = dataframe.cov()   #covraince 协方差
print(cov)
```

### 4.相关性矩阵

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pd.read_csv(filename, delimiter=',', header=None, names=names)

data_corr=dataframe.corr();

fig = plt.figure()

fig.set_size_inches(12,12)

ax = fig.add_subplot(111) 

cax = ax.matshow(data_corr, vmin=-1, vmax=1)
fig.colorbar(cax)

ticks  = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names,fontdict={'fontsize':20})
ax.set_yticklabels(names,fontdict={'fontsize':20})

plt.show()
```
